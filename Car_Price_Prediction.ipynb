```python
import datetime
date_time = datetime.datetime.now()
data['Age'] = date_time.year - data['Year']
data.head()
```

```python
data.drop('Year',axis=1,inplace=True);
data.head()
```

## **Outlier Removal**

```python
import seaborn as sns
sns.boxplot(data['Selling_Price'])
```

```python
sorted(data['Selling_Price'],reverse=True)
```

```python
data = data[~(data['Selling_Price']>=33.0) & (data['Selling_Price']<=35.0)]
```

```python
data.shape
```

```python
data.head(1)
```

```python
data['Fuel_Type'].unique()
```

```python
data['Fuel_Type']= data['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
```

```python
data['Fuel_Type'].unique()
```

```python
data['Selling_type'].unique()
```

```python
data['Selling_type']= data['Selling_type'].map({'Dealer':0,'Individual':1})
```

```python
data['Selling_type'].unique()
```

```python
data['Transmission'].unique()
```

```python
data['Transmission']= data['Transmission'].map({'Manual':0,'Automatic':1})
```

```python
data['Transmission'].unique()
```

```python
data.head()
```

```python
X=data.drop(['Car_Name','Selling_Price'],axis=1)
X.head()
```

```python
y=data['Selling_Price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

# **Import The models**

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

# **Model Training**

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
rf=RandomForestRegressor()
rf.fit(X_train,y_train)
gb=GradientBoostingRegressor()
gb.fit(X_train,y_train)
xgb=XGBRegressor()
xgb.fit(X_train,y_train)
```

# **Prediction on Test Data**

```python
y_pred1 = lr.predict(X_test)
y_pred2 = rf.predict(X_test)
y_pred3 = gb.predict(X_test)
y_pred4 = xgb.predict(X_test)
```

# **Evaluating the Algorithm**

```python
from sklearn import metrics
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
```

```python
print(score1)
print(score2)
print(score3)
print(score4)
```

```python
final_data=pd.DataFrame({'Models':['LR','RF','GBR','XG'],"R2_SCORE":[score1,score2,score3,score4]})
final_data
```

```python
import seaborn as sns
sns.barplot(data = final_data, x='Models',y='R2_SCORE')
```

# **Save The Model**

```python
xg = XGBRegressor()
xg_final = xg.fit(X,y)
```

```python
import joblib
joblib.dump(xg_final,'car_price_predictor')
```

```python
model = joblib.load('car_price_predictor')
```
